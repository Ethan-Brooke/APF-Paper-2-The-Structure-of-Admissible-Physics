## Display helper
A small utility to print theorem results consistently throughout the notebook.

````python
def show(result, extras=None):
    """Pretty-print a theorem check result dict."""
    ok   = result.get('passed', result.get('pass', '?'))
    name = result.get('name', '')
    ep   = result.get('epistemic', '')
    summ = result.get('summary', '')
    kr   = result.get('key_result', '')
    deps = result.get('dependencies', [])

    status_icon = '✓' if ok is True else ('✗' if ok is False else '?')
    status_text = 'PASS' if ok is True else ('FAIL' if ok is False else str(ok))
    color = '\033[92m' if ok is True else '\033[91m'
    reset = '\033[0m'

    print(f"{color}{status_icon} {status_text}{reset}  [{ep}]  {name}")
    print(f"   {summ[:110]}{'…' if len(summ)>110 else ''}")
    if kr:
        print(f"   → {kr[:110]}{'…' if len(kr)>110 else ''}")
    if extras:
        for k, v in extras.items():
            print(f"   {k}: {v}")
    if deps:
        dep_str = ', '.join(str(d) for d in deps[:8])
        if len(deps) > 8: dep_str += f' … (+{len(deps)-8} more)'
        print(f"   deps: {dep_str}")
    print()
````
